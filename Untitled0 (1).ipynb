```python
import matplotlib.pyplot as plt

# 결합확률질량함수 데이터
joint_pmf_data = {
    (1, 1): 0.1,
    (1, 2): 0.2,
    (1, 3): 0.2,
    (2, 1): 0.3,
    (2, 2): 0.05,
    (2, 3): 0.15
}

# X와 Y의 주변확률질량함수 계산
f_X = {}
f_Y = {}

for x, y in joint_pmf_data.keys():
    f_X[x] = f_X.get(x, 0) + joint_pmf_data[(x, y)]
    f_Y[y] = f_Y.get(y, 0) + joint_pmf_data[(x, y)]

# 그래프 범위 설정
x_min, x_max = 0.5, 3.5
y_min, y_max = 0, 1

# 그래프 크기 지정
plt.figure(figsize=(12, 5))

# -------------------------------
# X의 주변확률질량함수 그래프
# -------------------------------
plt.subplot(1, 2, 1)
plt.plot(list(f_X.keys()), list(f_X.values()), 'bo')  # 점
plt.vlines(list(f_X.keys()), y_min, list(f_X.values()), colors='blue')  # 수직선
plt.title('Marginal PMF of X')
plt.xlabel('X value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max) + 1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.grid(True)

# -------------------------------
# Y의 주변확률질량함수 그래프
# -------------------------------
plt.subplot(1, 2, 2)
plt.plot(list(f_Y.keys()), list(f_Y.values()), 'ro')  # 점
plt.vlines(list(f_Y.keys()), y_min, list(f_Y.values()), colors='red')  # 수직선
plt.title('Marginal PMF of Y')
plt.xlabel('Y value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max) + 1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# 국어 성적과 수학 성적 데이터
korean_scores = [75, 65, 50, 70, 80]
math_scores = [65, 70, 70, 55, 60]

# 데이터를 NumPy 배열로 변환
data = np.array([korean_scores, math_scores])

# 공분산 계산 (모집단 기준: bias=True)
covariance = np.cov(data, bias=True)[0, 1]

# 결과 출력
print("Covariance:", covariance)
```
